### merged

```python
df1 = specific_column_Saliva1_phylum
df2 = specific_column_Saliva2_phylum
df3 = specific_column_Saliva3_phylum
df4 = specific_column_Stool1_phylum
df5 = specific_column_Stool2_phylum

# merge
merged_phylum = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
merged_phylum

# select unique value
unique_specific_column_merged_phylum = set(merged_phylum[0])
len(unique_specific_column_merged_phylum)
```

##### x-axis

```python
# Saliva1 : 16

ctr_Saliva1_Archaea_phylum = collections.Counter()

for r in specific_column_Saliva1_phylum[0]:
    ctr_Saliva1_Archaea_phylum.update(r.split("/"))
print(ctr_Saliva1_Archaea_phylum)

df_ctr_Saliva1_phylum = pd.DataFrame(ctr_Saliva1_Archaea_phylum, index = ['Saliva1'])

# Saliva2 : 34

ctr_Saliva2_Archaea_phylum = collections.Counter()

for r in specific_column_Saliva2_phylum[0]:
    ctr_Saliva2_Archaea_phylum.update(r.split("/"))
print(ctr_Saliva2_Archaea_phylum)

df_ctr_Saliva2_phylum = pd.DataFrame(ctr_Saliva2_Archaea_phylum, index = ['Saliva2'])

# Saliva3 : 38

ctr_Saliva3_Archaea_phylum = collections.Counter()

for r in specific_column_Saliva3_phylum[0]:
    ctr_Saliva3_Archaea_phylum.update(r.split("/"))
print(ctr_Saliva3_Archaea_phylum)

df_ctr_Saliva3_phylum = pd.DataFrame(ctr_Saliva3_Archaea_phylum, index = ['Saliva3'])

# Stool1 : 41

ctr_Stool1_Archaea_phylum = collections.Counter()

for r in specific_column_Stool1_phylum[0]:
    ctr_Stool1_Archaea_phylum.update(r.split("/"))
print(ctr_Stool1_Archaea_phylum)

df_ctr_Stool1_phylum = pd.DataFrame(ctr_Stool1_Archaea_phylum, index = ['Stool1'])

# Stool2 : 37

ctr_Stool2_Archaea_phylum = collections.Counter()

for r in specific_column_Stool2_phylum[0]:
    ctr_Stool2_Archaea_phylum.update(r.split("/"))
print(ctr_Stool2_Archaea_phylum)

df_ctr_Stool2_phylum = pd.DataFrame(ctr_Stool2_Archaea_phylum, index = ['Stool2'])
```

- heatmap (list_saliva_bacteria)



```python
# 데이터프레임 생성

a = pd.DataFrame(unique_specific_column_Saliva1_phylum)
b = pd.DataFrame(unique_specific_column_Saliva2_phylum)
c = pd.DataFrame(unique_specific_column_Saliva3_phylum)
d = pd.DataFrame(unique_specific_column_Stool1_phylum)
e = pd.DataFrame(unique_specific_column_Stool2_phylum)

# 데이터프레임 병합하여 새로운 데이터프레임 생성
merged_df = pd.concat([a, b, c, d, e], keys=['a', 'b', 'c', 'd', 'e'])


# yticklabels 개수 설정
yticklabels_num = sum([len(df) for _, df in merged_df.groupby(level=0)])

# 그래프 크기 설정
fig = plt.figure(figsize=(10,30))
ax = fig.add_subplot(111)

# heatmap 
x_labels = list(unique_df_saliva_bacteria_species)
y_labels = list(merged_df.index.get_level_values(1))

data = np.random.rand(yticklabels_num, len(x_labels))
ax = sns.heatmap(data, xticklabels=x_labels, yticklabels=y_labels)

# 그래프 제목 설정
plt.title('<list_saliva_bacteria>')

# 그래프 y축 눈금 라벨 회전
plt.yticks(rotation=0)
plt.show

# save as jpg
plt.savefig("Compared_with_list_saliva_bacteria.jpg") 
```
